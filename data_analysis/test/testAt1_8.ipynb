```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
chi_data = pd.Series(np.random.chisquare(2,100000))
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
X,Y = pd.cut(chi_data,bins=50,retbins=True,precision=5)
```

```python
X.value_counts().sort_index()
```

```python
Y
```

```python
X.unique()
```

```python
ax.bar(Y[:50],X.value_counts().sort_index())
ax.plot(Y[:50],X.value_counts().sort_index())
fig
```

```python
t_data = pd.Series(np.random.standard_t(70,100000))
t_fig = plt.figure()
t_ax = t_fig.add_subplot(111)
bins=100
tX ,tXL = pd.cut(t_data,bins,retbins=True,precision=5)
t_ax.bar(tXL[:bins],tX.value_counts().sort_index())
t_ax.plot(tXL[:bins],tX.value_counts().sort_index(),color='orange')
```

```python
fsize = 100000
f_data = pd.Series(np.random.noncentral_f(17,90,1,size=fsize))
f_fig = plt.figure()
f_ax = f_fig.add_subplot(111)
bin_num = 100
fX,fbins = pd.cut(f_data,bin_num,retbins=True,precision=5)
bin_len = fbins[1] - fbins[0]
f_ax.hist(f_data,bin_num)
f_ax.plot(fbins[:bin_num],fX.value_counts().sort_index())
```

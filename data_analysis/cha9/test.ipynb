```python
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
data = pd.read_csv('examples/spx.csv')
```

```python
spx = data['SPX']
```

```python
date = data['Date'].apply(lambda x: datetime.date(datetime.strptime(x,'%Y-%m-%d %H:%M:%S')))
```

```python
date
```

```python
ax.clear()
ax.plot(date,spx,'b-')
ax.set_xlim(datetime(2007,1,1),datetime(2011,1,1))
ax.set_ylim(500,1800)
crisis_data = {
    datetime(2007,10,11):'Peak of bull Market',
    datetime(2008,3,12):'Bear Steams False',
    datetime(2008,9,15):'Lehman Bankruptcy'
}
for imp_date,label in crisis_data.items():
    ax.annotate(label,[imp_date,spx[data['Date'][data['Date']==datetime.strftime(imp_date,'%Y-%m-%d 00:00:00')].index]+75],
                xytext=(imp_date,spx[data['Date'][data['Date']==datetime.strftime(imp_date,'%Y-%m-%d 00:00:00')].index]+225),
                arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4),
                horizontalalignment='left',
                verticalalignment='top')
ax.set_title('Important Dates in the 2008-2009 financial crisis')
```

```python
fig
```

```python
t_data = pd.read_csv('examples/spx.csv',index_col=0,parse_dates=True)
```

```python
t_spx = t_data['SPX']
```

```python
t_spx
```

```python
t_fig = plt.figure()
t_ax = t_fig.add_subplot(111)
```

```python
t_ax.clear()
t_spx.plot(ax=t_ax,style='k-')
t_ax.set_xlim(datetime(2007,1,1),datetime(2011,1,1))
t_ax.set_ylim(500,1800)
for imp_date,events in crisis_data.items():
    t_ax.annotate(text=events,
                  xy=(imp_date,t_spx.asof(imp_date)+75),
                  xytext=(imp_date,t_spx.asof(imp_date)+225),
                  arrowprops=dict(
                      facecolor='black',
                      headwidth=4,
                      width=2,
                      headlength=4,
                      shrink=1
                      # arrowstyle='->',
                      # connectionstyle='angle,angleA=90,angleB=60,rad=10'
                      # connectionstyle=' angle,angleA=15,angleB=90'
                      ),
                horizontalalignment='left',
                verticalalignment='top'
                  )
t_ax.set_title('Important dates in 2008-2009 finacial crisis')
t_ax.set_xlabel('date in financial crirsis')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3,angle=-30,)
cric = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5)

ax.add_patch(rect)
ax.add_patch(cric)
ax.add_patch(pgon)
ax.plot(np.linspace(0,1,1000),[0.75]*1000)
ax.plot([0.2]*1000,np.linspace(0,1,1000))
```

```python
ax.clear()
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3,angle=-30)
cric = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3,)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6],[0.2,0]],color='g',alpha=0.5,closed=True)

ax.add_patch(rect)
ax.add_patch(cric)
ax.add_patch(pgon)
ax.plot(np.linspace(0,1,1000),[0.75]*1000)
ax.plot([0.2]*1000,np.linspace(0,1,1000))

ax.arrow(0.5,0.5,dx=0.3,dy=0.3,color='r')
```

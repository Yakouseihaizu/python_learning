```python
def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x]+=1
        else:
            counts[x] = 1
    return counts
```

```python
from collections import defaultdict
```

```python
def get_counts2(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] +=1
    return counts
```

```python
def get_counts3(sequence):
    counts = {}
    for x in sequence:
        counts.setdefault(x,0)
        counts[x]+=1
    return counts
```
